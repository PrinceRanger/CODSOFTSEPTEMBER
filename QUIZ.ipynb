# QUIZ GAME

```python
import random

# Define a list of quiz questions with answer
quiz_questions = [
    ("What is the capital of India?", ["A. Maharashtra", "B. Gujarat", "C. Delhi"], "C"),
    ("Which planet is known as the Red Planet?", ["A. Venus", "B. Mars", "C. Jupiter"], "B"),
    ("What is the largest mammal in the world?", ["A. Elephant", "B. Blue Whale", "C. Giraffe"], "B"),
    ("Which gas do plants absorb from the atmosphere?", ["A. Oxygen", "B. Nitrogen", "C. Carbon Dioxide"], "C"),
    ("Which sport is played on a court with a net and a shuttlecock?", ["A. Squash", "B. Badminton", "C. Volleyball"], "B"),
    ("Which country is the largest producer of coffee in the world?" , ["A. Brazil", "B. India", "C. Colombia"], "A"),
    ("Which country is the only one to border both the Atlantic Ocean and the Indian Ocean?" ,["A. South Africa", "B. Brazil", "C. Indonesia"], "A"),
    ("Which country has the most time zones in the world?" , ["A. Russia", "B. China", "C. France"], "C"),
    ("Which ancient civilization is credited with inventing the wheel?" , ["A. Sumerians", "B. Indians", "C. Egyptians"], "A"),
    ("Which ancient city was destroyed by a volcanic eruption in 79 AD?" ,["A. Athens", "B. Pompeii", "C. Babylon"], "B")
    
]

# starting message
def display_welcome_message():
    print("Welcome to the Quiz Game!")
    print("Answer the following multiple-choice questions.")

# question printing
def ask_question(question, options):
    print(question)
    for option in options:
        print(option)
    user_answer = input("Enter your answer (A/B/C): ").strip().upper()
    return user_answer

# evaluate answer
def evaluate_answer(user_answer, correct_answer):
    if user_answer == correct_answer:
        print("Correct!")
        return 1
    else:
        print(f"Incorrect. The correct answer is {correct_answer}.")
        return 0

# creating score
def main():
    display_welcome_message()
    score = 0

    for question, options, correct_answer in random.sample(quiz_questions, len(quiz_questions)):
        user_answer = ask_question(question, options)
        score += evaluate_answer(user_answer, correct_answer)

# printing the score
    print(f"Quiz Completed! Your score is {score}/{len(quiz_questions)}.")

if __name__ == "__main__":
    main()

```
